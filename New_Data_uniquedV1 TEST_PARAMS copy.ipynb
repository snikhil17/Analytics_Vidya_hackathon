```python
def mae(y_true, predictions):
    y_true, predictions = np.array(y_true), np.array(predictions)
    return np.mean(np.abs(y_true - predictions))
```

```python
def new_metric(df):
    df['mae'] = df[['quantity_received', 'preds']].apply(lambda row: mae(row['quantity_received'], row['preds']), axis=1)
    final_error = df['mae'].sum()/ df['preds'].sum()
    return final_error
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Define a function to calculate the VIF for each variable
def calc_vif(X):
    # Calculate VIF for each variable in the dataset
    vif = pd.DataFrame()
    vif["Features"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif

# Drop the target variable and any irrelevant features 
X = df_train[req_cols].copy()

# Calculate the initial VIF for each variable
initial_vif = calc_vif(X)

# Remove features with high VIF
while initial_vif['VIF'].max() > 5:
    # Get the name of the variable with the highest VIF
    remove_feature = initial_vif[initial_vif['VIF'] == initial_vif['VIF'].max()]['Features'].values[0]
    print("Removing Feature: ", remove_feature)
    # Drop the variable from the dataset
    X.drop(remove_feature, axis=1, inplace=True)
    # Recalculate the VIF for the remaining variables
    initial_vif = calc_vif(X)

# Print the final set of features after removing multicollinearity
print(X.columns)
```

```python
def mae(y_true, predictions):
    y_true, predictions = np.array(y_true), np.array(predictions)
    return np.mean(np.abs(y_true - predictions))
```

```python
def new_metric(df):
    df['mae'] = df[['quantity_received', 'preds']].apply(lambda row: mae(row['quantity_received'], row['preds']), axis=1)
    final_error = df['mae'].sum()/ df['preds'].sum()
    return final_error
```
